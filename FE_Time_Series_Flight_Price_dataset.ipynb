### Feature Engineering Process

```python
print(final_df['Date_of_Journey'].str.split('/').str[0],'\n',
      final_df['Date_of_Journey'].str.split('/').str[1],'\n',
      final_df['Date_of_Journey'].str.split('/').str[2])
```

### Spliting Date_of_Journey into Three seperate Columns (Year,Month,Day)

```python
df=final_df.copy()
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
# Another Smart Method
df['Date']=df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
df['Month']=df['Date_of_Journey'].apply(lambda x:x.split('/')[1])
df['Year']=df['Date_of_Journey'].apply(lambda x:x.split('/')[2])
```

```python
df[['Date','Month','Year']]=df[['Date','Month','Year']].astype('int64')
df.info()
```

### Dropping the Original Columns after splitting the data

```python
df.drop('Date_of_Journey',axis=1, inplace=True)
```

```python
df.head(10)
```

### Splitting Arrival Time into two seperate Hour and Minute Feature

```python
df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_Time']=df['Arrival_Time'].str.split(' ').str[0]
```

```python
df.head()
```

```python
print(df["Arrival_Time"].str.split(':').str[0],'\n',
      df["Arrival_Time"].str.split(':').str[1])      
```

```python
df["Arrival_Hour"]=df["Arrival_Time"].str.split(':').str[0]
df["Arrival_Minute"]=df["Arrival_Time"].str.split(':').str[1]
df.drop('Arrival_Time',axis=1,inplace=True)
df[['Arrival_Hour','Arrival_Minute']]=df[['Arrival_Hour','Arrival_Minute']].astype('int64')
display(df.info(),df.head())
```

### Splitting Departure Time into two seperate columns Hour and Minutes

```python
print(df["Dep_Time"].str.split(':').str[0],'\n',
      df["Dep_Time"].str.split(':').str[1]) 
```

### Convert the New columns Data type into Integer

```python
df["Departure_Hour"]=df["Dep_Time"].str.split(':').str[0]
df["Departure_Minute"]=df["Dep_Time"].str.split(':').str[1]
df.drop('Dep_Time',axis=1,inplace=True)
df[['Departure_Hour','Departure_Minute']]=df[['Departure_Hour','Departure_Minute']].astype('int64')
display(df.info(),df.head())
```

### Splitting Duration into Hour and Minute Column

```python
s=pd.to_timedelta(df['Duration'])
df['Duration_Hours']=s.dt.components['hours']
df['Duration_Minutes']=s.dt.components['minutes']
df.head()
```
